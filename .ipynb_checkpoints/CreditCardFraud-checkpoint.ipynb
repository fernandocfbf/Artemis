```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from mlxtend.plotting import plot_decision_regions
import numpy as np
from sklearn.decomposition import PCA
```

```python
dataset = pd.DataFrame(pd.read_csv('creditcard.csv'))
```

```python
dataset.head(3)
```

```python
#Checando as features do dataset
dataset.info()
```

```python
#Separando o dataset em treino e teste (Estou usando apenas mil linhas porque com 280000 o Jupyter crashava)
X_train, X_test, y_train, y_test = train_test_split(dataset.head(500), list(range(len(dataset.head(500)))), test_size=0.4)

#Aplicando o modelo SVM (Support Vector Machine)
model = svm.SVC(kernel = 'linear') #Kernel Linear

#Dando fit no modelo SVM usando o dataset de treino
model.fit(X_train, y_train)

#Testando o score do modelo
print(model.score(X_train, y_train))

#Caso vizualize esse score antes da limpeza do dataset, não estranhe, o score está alto devido ao desbalanceamento de casos de fraude,
#assim, o modelo quase sempre preve os casos teste como não sendo fraude (já que são a grande maioria) acertando em quase todos os casos.
```
