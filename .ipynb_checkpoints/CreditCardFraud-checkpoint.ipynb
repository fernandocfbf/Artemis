```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
#from mlxtend.plotting import plot_decision_regions
import numpy as np
from sklearn.decomposition import PCA
from sklearn.metrics import f1_score, recall_score
```

```python
dataset = pd.read_csv('creditcard.csv')
```

```python
dataset
```

Observando o valor em que ocorre a maioria das fraudes:

```python
fraud = dataset[dataset.Class == 1]
normal = dataset[dataset.Class == 0]

 

f, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
f.suptitle('Amount per transaction by class')
bins = 50
ax1.hist(fraud.Amount, bins = bins)
ax1.set_title('Fraud')
ax2.hist(normal.Amount, bins = bins)
ax2.set_title('Normal')
plt.xlabel('Amount ($)')
plt.ylabel('Number of Transactions')
plt.xlim((0, 20000))
plt.yscale('log')
plt.show();
```

Observando a diferença entre valores para cada categoria:

```python
columns = ["V1","V2","V3","V4","V5","V6","V7","V8","V9","V10","V11","V12","V13","V14","V15","V16","V17","V18","V19","V20","V21","V22","V23","V24","V25","V26","V27","V28"]
for e in columns:
    dataset.boxplot(column=e, by="Class")
    plt.title(f'{e} X Class')
    plt.ylabel(f"{e}")
    plt.xlabel('Class')
    plt.suptitle("")
    plt.show();
```

Observando a correlação entre valores:

```python
corrMatrix = dataset.corr()
index = corrMatrix.index
plt.figure(figsize=(20,20))
g=sns.heatmap(dataset[index].corr(),annot=True,cmap="RdYlGn")
```

Não é possível concluir, observando a  linha 'class' quais itens relamente possuem maior correlação. Isso pode ser explicado pela deisparidade, em termos de quantidade, entre fraudes e não fraudes. Por tanto, a diminuição e, desta forma, a paridade tornou - se o objetivo. 

```python
falseClass = dataset.loc[dataset.Class == 0]
trueClass = dataset.loc[dataset.Class == 1]
sizeTrue = trueClass.shape[0]
reducedFalse = falseClass.sample(sizeTrue)
```

```python
result = pd.concat([trueClass, reducedFalse], ignore_index=True, sort=True).sample(frac=1)
```

Abaixo pode ser observada a matriz de correlação quando a paridade numerica é presente:

```python
corrMatrix = result.corr()
index = corrMatrix.index
plt.figure(figsize=(20,20))
g=sns.heatmap(result[index].corr(),annot=True,cmap="RdYlGn")
```

Observando a coluna "class", pode-se concluir que diferentes grubos possuem um correlação maior que 0.5, em módulo, dessa forma, selecionou-se esse grupo.

```python
j = result.columns[abs(result[index].corr()['Class']) > 0.5][1::]
```

Apresenta-se abaixo a matriz de correlação entre os valores de maior correlação com 'class':

```python
corrMatrix = result.corr()
plt.figure(figsize=(20,20))
g=sns.heatmap(result[j].corr(),annot=True,cmap="RdYlGn")
```

Como pode ser observado, esse valores, além de possuir correlação com 'class' também posssuem grande correlação entre sí. Dessa forma, conclui-se a análise exploratória recomendando o uso das seguintes features: V3, V4, V9, V10, V11, V12, V14, V16, V17.

```python
dataset.head(3)
d = dataset.sort_values(by = 'Class', ascending = False)
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
X = d.iloc[0:1000, 0:30]
y = d.iloc[0:1000, 30:31]
```

```python
#Separando as bases de dados para teste e treino
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=RANDOM_SEED)

#Aplicando o modelo SVM (Support Vector Machine)
model = svm.SVC(kernel = 'linear') #Kernel Linear

#Dando fit no modelo SVM usando o dataset de treino
model.fit(X_train, y_train)

predict = model.predict(X_test)
```

```python
#Testando o score do modelo
model.score(X_test, y_test)
```

```python
#Resultados
print("O modelo SVM acertou {0}%".format(f1_score(y_test, predict) * 100))
```

```python
print("O modelo SVM acertou {0}%".format(recall_score(y_test, predict) * 100))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
dataset.columns
```

```python
# ------------- separando as colunas ------------
X =  dataset[['V3', 'V4', 'V9', 'V10', 'V11', 'V12', 'V14', 'V16', 'V17']]
y = dataset["Class"]
#X = dados.columns["]
```

```python
# Separação da base de treino e teste.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
mlp1 = MLPClassifier(hidden_layer_sizes = (440, 890, 36),alpha = 1e-6, max_iter = 1000, tol = 1e-6, random_state = 42, n_iter_no_change=40)
mlp1.fit(X_train, y_train.values.ravel())
```

```python
val_predictions = mlp1.predict(X_test)

f1_score(y_test, val_predictions, average='macro')
```
